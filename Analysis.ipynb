```python
sns.set(rc={'figure.figsize':(10,8)})
sns.barplot(x="Male", y="CAUSE", data=data)
plt.title("Causes v/s No. of Males dead in Road Accidents")
```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.barplot(x="Female", y="CAUSE", data=data)
plt.title("Causes v/s No. of females dead in Road Accidents")
```

```python
sns.barplot(x="Year", y="Total", data=data)
plt.title("Year v/s No. of people dead in Road Accidents")
```

```python
data_cor=data.corr()
sns.set(rc={'figure.figsize':(10,8)})
sns.heatmap(data_cor,annot=True)
```

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.lineplot(x="Year", y="Total",data=Maharasthra)
plt.title("No. of People dead every year in Maharasthra")
```

```python
sns.FacetGrid(Maharasthra, height=8).map(sns.distplot, "Male").add_legend()
plt.title("Distribution of males dead in Maharasthra")
```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.barplot(x="Male", y="CAUSE", data=Maharasthra)
plt.title("Causes v/s No. of males dead in Maharasthra Road Accidents")
```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.barplot(x="Female", y="CAUSE", data=Maharasthra)
plt.title("Causes v/s Number of males dead in Maharasthra Road Accidents")
```
